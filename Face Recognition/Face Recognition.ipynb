```python
obama_image = face_recognition.load_image_file("obama.jpg")
obama_face_encoding = face_recognition.face_encodings(obama_image)[0]

mark0_image = face_recognition.load_image_file("mark0.jpg")
mark0_face_encoding = face_recognition.face_encodings(mark0_image)[0]
mark1_image = face_recognition.load_image_file("mark1.jpg")
mark1_face_encoding = face_recognition.face_encodings(mark1_image)[0]

selena0_image = face_recognition.load_image_file("selena0.jpg")
selena0_face_encoding = face_recognition.face_encodings(selena0_image)[0]
selena1_image = face_recognition.load_image_file("selena1.jpg")
selena1_face_encoding = face_recognition.face_encodings(selena1_image)[0]

emma0_image = face_recognition.load_image_file("emma0.jpg")
emma0_face_encoding = face_recognition.face_encodings(emma0_image)[0]
emma1_image = face_recognition.load_image_file("emma1.jpg")
emma1_face_encoding = face_recognition.face_encodings(emma1_image)[0]

ahsan0_image = face_recognition.load_image_file("ahsan0.jpg")
ahsan0_face_encoding = face_recognition.face_encodings(ahsan0_image)[0]
ahsan1_image = face_recognition.load_image_file("ahsan1.jpg")
ahsan1_face_encoding = face_recognition.face_encodings(ahsan1_image)[0]
ahsan2_image = face_recognition.load_image_file("ahsan2.jpg")
ahsan2_face_encoding = face_recognition.face_encodings(ahsan2_image)[0]
ahsan3_image = face_recognition.load_image_file("ahsan3.jpg")
ahsan3_face_encoding = face_recognition.face_encodings(ahsan3_image)[0]

raheela0_image = face_recognition.load_image_file("raheela0.jpg")
raheela0_face_encoding = face_recognition.face_encodings(raheela0_image)[0]
raheela1_image = face_recognition.load_image_file("raheela1.jpg")
raheela1_face_encoding = face_recognition.face_encodings(raheela1_image)[0]
raheela2_image = face_recognition.load_image_file("raheela2.jpg")
raheela2_face_encoding = face_recognition.face_encodings(raheela2_image)[0]
raheela3_image = face_recognition.load_image_file("raheela3.jpg")
raheela3_face_encoding = face_recognition.face_encodings(raheela3_image)[0]
raheela4_image = face_recognition.load_image_file("raheela4.jpg")
raheela4_face_encoding = face_recognition.face_encodings(raheela4_image)[0]
```

```python
known_face_encodings = [
    obama_face_encoding,
    mark0_face_encoding,
    mark1_face_encoding,
    selena0_face_encoding,
    selena1_face_encoding,
    emma0_face_encoding,
    emma1_face_encoding,
    ahsan0_face_encoding,
    ahsan1_face_encoding,
    ahsan2_face_encoding,
    ahsan3_face_encoding,
    raheela0_face_encoding,
    raheela1_face_encoding,
    raheela2_face_encoding,
    raheela3_face_encoding,
    raheela4_face_encoding,
]

known_face_names = [
    "Barack Obama",
    "Mark Zunckerberg",
    "Mark Zunckerberg",
    "Selena Gomez",
    "Selena Gomez",
    "Emma Watson",
    "Emma Watson",
    "Ahsan Jawaid",
    "Ahsan jawaid",
    "Ahsan Jawaid",
    "Ahsan jawaid",
    "Raheela Bano",
    "Raheela Bano",
    "Raheela Bano",
    "Raheela Bano",
    "Raheela Bano",
]
```

```python
face_locations = []
face_encodings = []
face_names = []
process_this_frame = True
```

```python
while True:
    ret, frame = video_capture.read()

    small_frame = cv2.resize(frame, (0, 0), fx=0.25, fy=0.25)

    rgb_small_frame = small_frame[:, :, ::-1]

    if process_this_frame:
        face_locations = face_recognition.face_locations(rgb_small_frame)
        face_encodings = face_recognition.face_encodings(rgb_small_frame, face_locations)

        face_names = []
        for face_encoding in face_encodings:
            matches = face_recognition.compare_faces(known_face_encodings, face_encoding)
            name = "Unknown"

            # # If a match was found in known_face_encodings, just use the first one.
            # if True in matches:
            #     first_match_index = matches.index(True)
            #     name = known_face_names[first_match_index]

            # Or instead, use the known face with the smallest distance to the new face
            face_distances = face_recognition.face_distance(known_face_encodings, face_encoding)
            best_match_index = np.argmin(face_distances)
            if matches[best_match_index]:
                name = known_face_names[best_match_index]

            face_names.append(name)

    process_this_frame = not process_this_frame


    for (top, right, bottom, left), name in zip(face_locations, face_names):
        top *= 4
        right *= 4
        bottom *= 4
        left *= 4

        cv2.rectangle(frame, (left, top), (right, bottom), (0, 0, 255), 2)

        cv2.rectangle(frame, (left, bottom - 35), (right, bottom), (0, 0, 255), cv2.FILLED)
        font = cv2.FONT_HERSHEY_DUPLEX
        cv2.putText(frame, name, (left + 6, bottom - 6), font, 1.0, (255, 255, 255), 1)

    cv2.imshow('Video', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video_capture.release()
cv2.destroyAllWindows()
```
